# Example of usage

```python
import matplotlib.pyplot as plt
import torch
```

Generate Dataset: `utils.genDataset`

```python
from utils.genDataset import synthetic_data
```

```python
w = torch.Tensor([2, -3.4])
b = 4.2
num_examples = 1000
features, labels = synthetic_data(w, b, num_examples)
```

```python
plt.scatter(features[:, 1].detach().numpy(),
            labels.detach().numpy(), 1)
```

Read Dataset: `utils.readDataset`

```python
from utils.readDataset import data_iter
```

```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```
